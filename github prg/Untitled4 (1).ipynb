```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:/Users/Mathematics/OneDrive/Desktop/dataset/tvmarketing.csv")
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
target_y=df["Sales"]
features_x=df.drop('Sales', axis=1, inplace=False)
features_x.head()
```

```python
target_y=df["Advertisments"]
features_x=df.drop('Advertisments', axis=1, inplace=False)
features_x.head()
```

```python
train_x, test_x, train_y, test_y=train_test_split(features_x, target_y, test_size=0.8)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(train_x,train_y)
```

```python
print('co-efficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
predicted_y=model.predict(test_x)
predicted_y.shape
```

```python
from sklearn.metrics import r2_score
print("accuracy:",r2_score(test_y, predicted_y))
```
