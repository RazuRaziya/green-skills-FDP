```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.decomposition import PCA
```

```python
iris=datasets.load_iris()
iris
```

```python
df=pd.DataFrame(iris['data'],columns=iris['feature_names'],)
df
```

```python
iris.target
```

```python
scalar=StandardScaler()
scaled_data=pd.DataFrame(scalar.fit_transform(df))
scaled_data
                
```

```python
sns.heatmap(scaled_data.corr(),
```

```python
pca=PCA(n_components=4)
pca.fit(scaled_data)
data_pca=pca.transform(scaled_data)
data_pca=pd.DataFrame(data_pca,columns=['PC1','PC2','PC3','PC4'])
data_pca.head()
```

```python
sns.scatterplot(x=data_pca["PC1"], y=data_pca["PC2"], hue=iris["target"])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
```

```python
faces = fetch_olivetti_faces(shuffle=True, random_state=42)
X = faces.data  # Shape: (400, 4096) - 400 images, each with 64x64 pixels
original_image = X[0].reshape(64, 64)

plt.figure(figsize=(4, 4))
plt.imshow(original_image, cmap='gray')
plt.title("Original Image (64x64 = 4096 pixels)")
plt.axis('off')
plt.show()

print(f"🔹 Original Image Shape: {original_image.shape} = {original_image.size} pixels")
```

```python
pca_50 = PCA(n_components=50, whiten=True, random_state=42)
pca_100 = PCA(n_components=100, whiten=True, random_state=42)
pca_200 = PCA(n_components=200, whiten=True, random_state=42)

X_pca_50 = pca_50.fit_transform(X)   # Shape: (400, 50)
X_pca_100 = pca_100.fit_transform(X) # Shape: (400, 100)
X_pca_200 = pca_200.fit_transform(X) # Shape: (400, 200)
X_reconstructed_50 = pca_50.inverse_transform(X_pca_50)
X_reconstructed_100 = pca_100.inverse_transform(X_pca_100)
X_reconstructed_200 = pca_200.inverse_transform(X_pca_200)
def plot_images(original, reconstructed, title, size=(64, 64)):
    plt.figure(figsize=(8, 4))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(original.reshape(size), cmap='gray')
    plt.title("Original Image (4096 pixels)")
    plt.axis('off')

    # Reconstructed image
    plt.subplot(1, 2, 2)
    plt.imshow(reconstructed.reshape(size), cmap='gray')
    plt.title(title)
    plt.axis('off')

    plt.show()

# 📌 Show Reconstruction Quality at Different PCA Levels
plot_images(X[0], X_reconstructed_50[0], "Reconstructed (50 Components)")
print(f"🔹 With 50 components, shape = {X_pca_50.shape[1]} (vs. 4096)")

plot_images(X[0], X_reconstructed_100[0], "Reconstructed (100 Components)")
print(f"🔹 With 100 components, shape = {X_pca_100.shape[1]} (vs. 4096)")

plot_images(X[0], X_reconstructed_200[0], "Reconstructed (200 Components)")
print(f"🔹 With 200 components, shape = {X_pca_200.shape[1]} (vs. 4096)")
```
